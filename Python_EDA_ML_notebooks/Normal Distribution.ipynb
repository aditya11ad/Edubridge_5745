# Noraml Distribution:

```python
import numpy as np
data= np.random.normal(size =1000,loc= 25,scale =7 )
```

```python
import seaborn as sns
sns.distplot(data)
```

# Skewed Distribution:


```python
from scipy.stats import skewnorm
```

### Right Skewed:

```python
data_r= skewnorm.rvs(10, size=1000,loc= 25,scale =7)
sns.distplot(data_r)
```

```python
[np.median(data_r) ,np.mean(data_r)]
```

```python
sns.boxplot(data_r)
```

### Left Skewed:

```python
data_l= skewnorm.rvs(-10, size=1000,loc= 25,scale =7)
sns.distplot(data_l)
```

```python
[np.median(data_l) , np.mean(data_l)]
```

```python
sns.boxplot(data_l)
```
